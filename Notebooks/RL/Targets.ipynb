```python
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
rewards = pd.Series(np.random.randn(20000)/1000)
rewards.plot();
```

```python
(1+rewards).cumprod().plot();
```

```python
reversed_rewards = np.flip(rewards, axis=0)
reversed_cum_ret = np.cumprod(1 + reversed_rewards)
cum_ret_t = np.flip(reversed_cum_ret, axis=0)
(cum_ret_t - 1).plot();
```

```python
gamma = 0.9999
trajsize = rewards.size
gammas = np.power(gamma, np.arange(trajsize))
cum_ret_t = pd.Series(np.zeros(trajsize))
for t in range(trajsize):
    if t == 0:
        cum_ret_t[t] = np.cumprod(1 + rewards*gammas).values[-1]
    else:
        cum_ret_t[t] = np.cumprod(1 + rewards[t:] * gammas[:-t]).values[-1]
(cum_ret_t - 1).plot();        
```

```python
pd.Series(np.power(0.9999, np.arange(trajsize))).plot();
```
